# Smart Lender - Data Preprocessing
- STEP 1 : Remove Loan_ID
- STEP 2 : Check Missing Values
- STEP 3 : Replace Missing Values
- STEP 4 : Encode Categorical Variables
- STEP 5 : Handle Outliers
- STEP 6 : Split Features and Target
- STEP 7 : Train Test Split
- STEP 8 : SMOTE
- STEP 9 : Feature Scaling
- STEP 10 : Save Encoders & Scaler

### STEP 1 : Remove Loan_ID

```python
# Remove Loan_ID column

df.drop("Loan_ID", axis=1, inplace=True)

# Display first 5 rows

df.head()
```

### STEP 2 : Check Missing Values

```python
# Display missing values in each column

print(df.isnull().sum())
```

### STEP 3 : Replace Missing Values

```python
# ===============================
# Fill Missing Values
# ===============================

# Categorical Columns → Mode

df["Gender"] = df["Gender"].fillna(df["Gender"].mode()[0])

df["Married"] = df["Married"].fillna(df["Married"].mode()[0])

df["Dependents"] = df["Dependents"].fillna(df["Dependents"].mode()[0])

df["Self_Employed"] = df["Self_Employed"].fillna(df["Self_Employed"].mode()[0])

df["Loan_Amount_Term"] = df["Loan_Amount_Term"].fillna(df["Loan_Amount_Term"].mode()[0])

df["Credit_History"] = df["Credit_History"].fillna(df["Credit_History"].mode()[0])

# Numerical Column → Median

df["LoanAmount"] = df["LoanAmount"].fillna(df["LoanAmount"].median())

print(df.isnull().sum())
```

### STEP 4 : Encode Categorical Variables

```python
from sklearn.preprocessing import LabelEncoder

# Dictionary to store encoders
encoders = {}

categorical_columns = [
    "Gender",
    "Married",
    "Education",
    "Self_Employed",
    "Property_Area",
    "Loan_Status"
]

for column in categorical_columns:

    le = LabelEncoder()

    df[column] = le.fit_transform(df[column])

    encoders[column] = le

# Convert Dependents

df["Dependents"] = df["Dependents"].replace("3+", 3)

df["Dependents"] = df["Dependents"].astype(int)

df.head()
```

### STEP 5 : Handle Outliers

```python
# Numerical Columns

numerical_columns = [

    "ApplicantIncome",

    "CoapplicantIncome",

    "LoanAmount"

]

for column in numerical_columns:

    Q1 = df[column].quantile(0.25)

    Q3 = df[column].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR

    upper = Q3 + 1.5 * IQR

    df = df[(df[column] >= lower) &
            (df[column] <= upper)]

print(df.shape)
```

### STEP 6 : Split Features and Target

```python
# Independent variables

X = df.drop("Loan_Status", axis=1)

# Dependent variable

y = df["Loan_Status"]

print(X.shape)
print(y.shape)
```

### STEP 7 : Train Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(

    X,

    y,

    test_size=0.20,

    random_state=42,

    stratify=y

)

print(X_train.shape)

print(X_test.shape)
```

### STEP 8 : SMOTE

```python
from imblearn.over_sampling import SMOTE

# Create SMOTE object

smote = SMOTE(random_state=42)

# Apply ONLY on training data

X_train, y_train = smote.fit_resample(

    X_train,

    y_train

)

print(X_train.shape)

print(y_train.shape)
```

### STEP 9 : Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

# Create scaler object

scaler = StandardScaler()

# Fit only on training data
# Transform both training and testing data

X_train = scaler.fit_transform(X_train)

X_test = scaler.transform(X_test)

print(X_train)
```

### STEP 10 : Save Encoders & Scaler

```python
import pickle

# Save scaler

with open("../model/scaler.pkl", "wb") as file:

    pickle.dump(scaler, file)

# Save all encoders

with open("../model/encoders.pkl", "wb") as file:

    pickle.dump(encoders, file)

print("Scaler and Encoders saved successfully.")
```

### Preprocessing Done

# Machine Learning Model Training

```python
# ============================================
# Store performance of all models
# ============================================

model_names = []

training_accuracy = []

testing_accuracy = []

cross_validation_score = []
```

## MODEL 1 : Decision Tree

```python
# ============================================
# Import Decision Tree
# ============================================

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.model_selection import cross_val_score

# ============================================
# Create Decision Tree Model
# ============================================

dt_model = DecisionTreeClassifier(random_state=42)

# ============================================
# Train Model
# ============================================

dt_model.fit(X_train, y_train)

# ============================================
# Predictions
# ============================================

train_prediction = dt_model.predict(X_train)
test_prediction = dt_model.predict(X_test)

# ============================================
# Accuracy
# ============================================

train_acc = accuracy_score(y_train, train_prediction)
test_acc = accuracy_score(y_test, test_prediction)

print("Training Accuracy :", train_acc)
print("Testing Accuracy  :", test_acc)

# ============================================
# Confusion Matrix
# ============================================

cm = confusion_matrix(y_test, test_prediction)

print("\nConfusion Matrix")
print(cm)

# ============================================
# Classification Report
# ============================================

print("\nClassification Report")
print(classification_report(y_test, test_prediction))

# ============================================
# Cross Validation
# ============================================

cv_score = cross_val_score(
    DecisionTreeClassifier(random_state=42),
    X,
    y,
    cv=5,
    scoring="accuracy"
)

print("Cross Validation Scores :", cv_score)
print("Average CV Score :", cv_score.mean())

# ============================================
# Save Result
# ============================================

model_names.append("Decision Tree")
training_accuracy.append(train_acc)
testing_accuracy.append(test_acc)
cross_validation_score.append(cv_score.mean())
```

### Result Table

```python
result = pd.DataFrame({

    "Model": model_names,

    "Training Accuracy": training_accuracy,

    "Testing Accuracy": testing_accuracy,

    "Cross Validation": cross_validation_score

})

result
```

# MODEL 4 : XGBoost

```python
# ============================================
# Import XGBoost
# ============================================

from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.model_selection import cross_val_score

# ============================================
# Create XGBoost Model
# ============================================

xgb_model = XGBClassifier(
    random_state=42,
    eval_metric="logloss"
)

# ============================================
# Train Model
# ============================================

xgb_model.fit(X_train, y_train)

# ============================================
# Predictions
# ============================================

train_prediction = xgb_model.predict(X_train)
test_prediction = xgb_model.predict(X_test)

# ============================================
# Accuracy
# ============================================

train_acc = accuracy_score(y_train, train_prediction)
test_acc = accuracy_score(y_test, test_prediction)

print("Training Accuracy :", train_acc)
print("Testing Accuracy  :", test_acc)

# ============================================
# Confusion Matrix
# ============================================

cm = confusion_matrix(y_test, test_prediction)

print("\nConfusion Matrix")
print(cm)

# ============================================
# Classification Report
# ============================================

print("\nClassification Report")
print(classification_report(y_test, test_prediction))

# ============================================
# Cross Validation
# ============================================

cv_score = cross_val_score(
    XGBClassifier(
    n_estimators=300,
    max_depth=5,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    eval_metric="logloss"
),
    X,
    y,
    cv=5,
    scoring="accuracy"
)

print("Cross Validation Scores :", cv_score)
print("Average CV Score :", cv_score.mean())

# ============================================
# Save Result
# ============================================

model_names.append("XGBoost")
training_accuracy.append(train_acc)
testing_accuracy.append(test_acc)
cross_validation_score.append(cv_score.mean())
```
